If we look at the `type` of the time columns we can see that it's a unicode string, not a date time. If the columns were named differently then it'd be fine (See [`convert_dates` here](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_json.html)). As it's not we need to convert it.

```python
print data1.time[0]
print type(data1.time[0])
```

```python
def convert_timestring(row):
    return dateutil.parser.parse(row.time)
    

data1["timestamp"] = data1.apply (lambda row: convert_timestring(row), axis=1)
```

This looks hardcore, but it just goes down the table making a new cell in each row that's formatted as a date not a string.

Now when we check again it shows that we've got a `pandas.tslib.Timestamp` object 

```python
print data1.timestamp[0]
print type(data1.timestamp[0])
```

Here's a simple plot. We can see that there's a trend here. So now we can start to break it down a bit more.

```python
plt.plot_date(data1.timestamp, data1.rssi)
```

```python
#create unique list of names
uniqueNames = data1.major.unique()

#create a data frame dictionary to store your data frames
DataFrameDict = {elem : pandas.DataFrame for elem in uniqueNames}

for key in DataFrameDict.keys():
    DataFrameDict[key] = data1[:][data1.major == key] 
    
print uniqueNames
print DataFrameDict[5].shape
print DataFrameDict[9752].shape
```

```python
hockey = DataFrameDict[5]
estimote = DataFrameDict[9752]
plt.plot_date(estimote.timestamp, estimote.rssi, label='Estimote')
plt.plot_date(hockey.timestamp, hockey.rssi, label='Hockey')
plt.xlabel('Time')
plt.ylabel('RSSI')
plt.title('RSSI Vs Time')
plt.legend()
plt.show()
```

So we can see that one of the beacons had a pretty consistently stronger signal. I think it was the hockey puck one. _However_, they both follow the same shape.

The following try is based on [this](http://stackoverflow.com/questions/16716302/how-do-i-fit-a-sine-curve-to-my-data-with-pylab-and-numpy) but just a manual guess at a fit. I'll figure out the scipy stuff when we have more data.

```python
import numpy as np

N = 1000 # number of data points
t = np.linspace(0, 4*np.pi, N)
# data = 3.0*np.sin(t+0.001) + 0.5 + np.random.randn(N) # create artificial data with noise

rssiShift = -2.5
pitchSpread = 0.9
guess_phase = np.deg2rad(300) # just twiddle these numbers until it looks right
guess_mean = np.mean(hockey.rssi)
guess_std = 3*np.std(hockey.rssi)/(2**0.5)

print guess_mean, guess_std, guess_phase

# we'll use this to plot our first estimate. This might already be good enough for you
data_first_guess = guess_std*np.sin((t+guess_phase)* pitchSpread) + guess_mean + rssiShift

plt.plot(estimote.rssi, "." , label='RSSI, estimote')
plt.plot(hockey.rssi, "." , label='RSSI, hockey')
plt.plot(data_first_guess, label='first guess')
plt.legend()
plt.show()
```

This is extremely rough, but it seems to show that there's something to think about here. If we had more turns then we could see if this sine wave persists.

The curve is only fitted to the hockey data. It looks like it would be the same curve, just weaker for the estimote data.
